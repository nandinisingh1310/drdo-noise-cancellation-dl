```python
import os
import librosa
import numpy as np
import soundfile as sf
from scipy.signal import butter, lfilter

# Band / High / Low Pass Filters
def butter_filter(lowcut=None, highcut=None, fs=16000, order=4, btype='band'):
    nyq = 0.5 * fs
    if btype == 'band':
        b, a = butter(order, [lowcut/nyq, highcut/nyq], btype='band')
    elif btype == 'high':
        b, a = butter(order, lowcut/nyq, btype='high')
    elif btype == 'low':
        b, a = butter(order, highcut/nyq, btype='low')
    return b, a

def apply_filter(audio, lowcut=None, highcut=None, fs=16000, order=4, btype='band'):
    b, a = butter_filter(lowcut, highcut, fs=fs, order=order, btype=btype)
    return lfilter(b, a, audio)

def normalize(audio):
    return audio / np.max(np.abs(audio))
```

```python
def filter_folder(input_folder, output_folder, filter_type=None, lowcut=None, highcut=None, sr=16000):
    os.makedirs(output_folder, exist_ok=True)

    for file in os.listdir(input_folder):
        if file.endswith(".wav"):
            path = os.path.join(input_folder, file)
            audio, _ = librosa.load(path, sr=sr)

            # Apply Filter
            if filter_type:
                audio = apply_filter(audio, lowcut=lowcut, highcut=highcut, fs=sr, btype=filter_type)

            # Normalize
            audio = normalize(audio)

            # Save
            sf.write(os.path.join(output_folder, file), audio, sr)
            print(f"Saved: {file}")
```

```python
import os
import random
import librosa
import numpy as np
import soundfile as sf
from glob import glob


clean_folder = "/content/drive/MyDrive/noise_project/clean_avg_fixed"
static_folder = "/content/drive/MyDrive/noise_project/length_fixed/static"
dynamic_folder = "/content/drive/MyDrive/noise_project/length_fixed"
output_folder = "/content/drive/MyDrive/mixed_dataset/noisy"
os.makedirs(output_folder, exist_ok=True)


dynamic_types = ["wind", "gun", "helicopter", "explosion"]


def load_audio(file, length, sr=16000):
    y, _ = librosa.load(file, sr=sr)
    y = y / np.max(np.abs(y))  # normalize between -1 and 1
    if len(y) < length:
        y = np.pad(y, (0, length - len(y)))
    else:
        y = y[:length]
    return y


def mix_audio(clean_file):
    clean, sr = librosa.load(clean_file, sr=16000)
    clean = clean / np.max(np.abs(clean))  # normalize
    length = len(clean)

    # 1. Pick one random static noise
    static_files = glob(os.path.join(static_folder, "*.wav"))
    static_noise = load_audio(random.choice(static_files), length)
    static_noise *= 0.1  # low volume

    # 2. Pick one random dynamic type + file
    chosen_type = random.choice(dynamic_types)
    dynamic_files = glob(os.path.join(dynamic_folder, chosen_type, "*.wav"))
    dynamic_noise = load_audio(random.choice(dynamic_files), length)
    dynamic_noise *= 0.25  # ower than clean

    # 3. Mix all three signals
    mixed = clean + static_noise + dynamic_noise
    mixed = mixed / np.max(np.abs(mixed))  # normalize again

    # 4. Save with name
    name = os.path.basename(clean_file).replace(".wav", "")
    new_name = f"{name}_{chosen_type}_radio.wav"
    save_path = os.path.join(output_folder, new_name)
    sf.write(save_path, mixed, sr)
    print(" Mixed & Saved:", new_name)

# === Run mixing ===
clean_files = glob(os.path.join(clean_folder, "*.wav"))
print(f"Total clean files found: {len(clean_files)}")

for file in clean_files:
    mix_audio(file)
```

```python
import os
import numpy as np
import librosa
from glob import glob

# ==== CONFIG ====
sr = 16000
fixed_length = 48800  # 2 sec audio

# ==== PATHS (CHANGE IF NEEDED) ====
clean_dir = "/content/drive/MyDrive/noise_project/clean_avg_fixed"
noisy_dir = "/content/drive/MyDrive/noise_project/mixed_dataset/noisy"

# ==== LOAD + PAD FUNCTION ====
def load_and_pad(file, length=fixed_length):
    audio, _ = librosa.load(file, sr=sr)
    audio = audio / np.max(np.abs(audio))
    if len(audio) < length:
        audio = np.pad(audio, (0, length - len(audio)))
    return audio[:length]

# ==== GET FILE LISTS ====
clean_files = sorted(glob(os.path.join(clean_dir, "*.wav")))
noisy_files = sorted(glob(os.path.join(noisy_dir, "*.wav")))

print(f"Found {len(clean_files)} clean and {len(noisy_files)} noisy files")

# ==== SAFETY CHECK ====
min_len = min(len(clean_files), len(noisy_files))
clean_files = clean_files[:min_len]
noisy_files = noisy_files[:min_len]

# ==== LOAD DATA ====
X = []
Y = []

for clean_f, noisy_f in zip(clean_files, noisy_files):
    clean_audio = load_and_pad(clean_f)
    noisy_audio = load_and_pad(noisy_f)
    X.append(noisy_audio)
    Y.append(clean_audio)

X = np.array(X)
Y = np.array(Y)

print("✅ Loaded data:")
print("  X (noisy):", X.shape)
print("  Y (clean):", Y.shape)
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)


X_train = X_train[..., np.newaxis]
Y_train = Y_train[..., np.newaxis]
X_val = X_val[..., np.newaxis]
Y_val = Y_val[..., np.newaxis]
```

```python
from tensorflow.keras import layers, models

def build_1d_cae(input_shape):
    model = models.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv1D(16, 9, padding="same", activation="relu"),
        layers.MaxPooling1D(2),
        layers.Conv1D(32, 9, padding="same", activation="relu"),
        layers.MaxPooling1D(2),
        layers.Conv1D(64, 9, padding="same", activation="relu"),

        layers.Conv1DTranspose(64, 9, padding="same", activation="relu"),
        layers.UpSampling1D(2),
        layers.Conv1DTranspose(32, 9, padding="same", activation="relu"),
        layers.UpSampling1D(2),
        layers.Conv1D(1, 9, padding="same", activation="tanh")
    ])
    return model

model = build_1d_cae((fixed_length, 1))
model.compile(optimizer="adam", loss="mse")
model.summary()
```

```python
history = model.fit(
    X_train, Y_train,
    validation_data=(X_val, Y_val),
    epochs=10,
    batch_size=16
)
```

```python
from tensorflow.keras import layers, models
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, UpSampling1D, LSTM, LeakyReLU

def build_cnn_lstm_denoiser(input_shape):
    inputs = Input(shape=input_shape)

    # Encoder
    x = Conv1D(32, kernel_size=9, padding="same")(inputs)
    x = LeakyReLU(alpha=0.01)(x)
    x = MaxPooling1D(pool_size=2)(x)

    x = Conv1D(48, kernel_size=9, padding="same")(x)
    x = LeakyReLU(alpha=0.01)(x)
    x = MaxPooling1D(pool_size=2)(x)

    # LSTM Layers
    x = LSTM(128, return_sequences=True)(x)
    x = LSTM(64, return_sequences=True)(x)

    # Decoder
    x = UpSampling1D(size=2)(x)
    x = UpSampling1D(size=2)(x)

    outputs = Conv1D(1, kernel_size=9, padding="same", activation="tanh")(x)

    model = models.Model(inputs, outputs)
    return model

# Define input shape
input_shape = (48800, 1)

# Build and compile model
model = build_cnn_lstm_denoiser(input_shape)
model.compile(optimizer="rmsprop", loss="mse")
model.summary()

# Train the model
history = model.fit(
    X_train, Y_train,
    validation_data=(X_val, Y_val),
    epochs=22,
    batch_size=16
)
```

```python
import os
import librosa
import numpy as np

# === Parameters
sr = 16000
duration = 2
samples = sr * duration
n_mels = 128
n_fft = 1024
hop_length = 512

# === Paths
noisy_dir = "/content/drive/MyDrive/noise_project/mixed_dataset/noisy"
clean_dir = "/content/drive/MyDrive/noise_project/clean_avg_fixed"

# === Function to convert audio to normalized log-mel spectrogram
def to_mel_spec(path):
    y, _ = librosa.load(path, sr=sr)
    y = librosa.util.fix_length(y, size=samples)
    y = y / np.max(np.abs(y))
    mel = librosa.feature.melspectrogram(y, sr=sr, n_fft=n_fft,
                                         hop_length=hop_length, n_mels=n_mels)
    mel_db = librosa.power_to_db(mel, ref=np.max)
    mel_db_norm = (mel_db + 80) / 80
    return mel_db_norm

# === Load all file pairs
X, Y = [], []
file_list = sorted(os.listdir(noisy_dir))
for file in file_list:
    if file.endswith(".wav"):
        noisy_path = os.path.join(noisy_dir, file)
        clean_path = os.path.join(clean_dir, file)
        if os.path.exists(clean_path):
            X.append(to_mel_spec(noisy_path))
            Y.append(to_mel_spec(clean_path))
        else:
            print(f"❌ Clean file missing for: {file}")

# === Convert to arrays for training
X = np.array(X)[..., np.newaxis]  # shape: (N, 128, T, 1)
Y = np.array(Y)[..., np.newaxis]
print("✅ Dataset ready:", X.shape, Y.shape)
```
